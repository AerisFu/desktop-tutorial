## 傅里叶基的数学理论与图像展示

### 数学理论
傅里叶变换是一种将信号从时域转换为频域的数学工具。它基于这样一个思想：任何周期性函数都可以表示为正弦和余弦函数的线性组合。这些正弦和余弦函数构成了傅里叶变换的基函数，称为**傅里叶基**。

#### 连续傅里叶变换
对于一个连续函数 \( f(t) \)，其傅里叶变换为：
\[
F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-i \omega t} \, dt
\]
其中 \( \omega = 2\pi f \) 是角频率。

逆变换为：
\[
f(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} F(\omega) e^{i \omega t} \, d\omega
\]

#### 离散傅里叶变换 (DFT)
对于离散信号 \( x[n] \) (n = 0, 1, ..., N-1)，DFT 为：
\[
X[k] = \sum_{n=0}^{N-1} x[n] e^{-i 2\pi k n / N}
\]
其中 k = 0, 1, ..., N-1。

基函数是 \( e^{i 2\pi k n / N} \)，这是复指数函数。

在实数域中，可以分解为正弦和余弦：
\[
e^{i \theta} = \cos \theta + i \sin \theta
\]
所以 DFT 可以写成实部和虚部，对应于余弦和正弦分量。

#### 傅里叶基的性质
- **正交性**：基函数在适当的内积下正交。
- **完备性**：任何函数都可以用这些基函数表示。
- **频率分辨率**：基函数的频率由 k/N 决定。

下面我们将通过图像展示这些基函数。

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置参数
N = 64  # 离散点数
n = np.arange(N)
k_values = [0, 1, 2, 4, 8]  # 不同的频率索引

# 创建子图
fig, axes = plt.subplots(len(k_values), 3, figsize=(15, 12))

for i, k in enumerate(k_values):
    # 复指数基函数 (实部和虚部)
    basis_real = np.cos(2 * np.pi * k * n / N)
    basis_imag = np.sin(2 * np.pi * k * n / N)
    basis_complex = np.exp(1j * 2 * np.pi * k * n / N)
    
    # 绘制实部 (余弦)
    axes[i, 0].plot(n, basis_real, 'b-', linewidth=2)
    axes[i, 0].set_title(f'余弦基: k={k} (实部)')
    axes[i, 0].set_xlabel('n')
    axes[i, 0].set_ylabel('幅度')
    axes[i, 0].grid(True)
    
    # 绘制虚部 (正弦)
    axes[i, 1].plot(n, basis_imag, 'r-', linewidth=2)
    axes[i, 1].set_title(f'正弦基: k={k} (虚部)')
    axes[i, 1].set_xlabel('n')
    axes[i, 1].set_ylabel('幅度')
    axes[i, 1].grid(True)
    
    # 绘制复数基的实部和虚部
    axes[i, 2].plot(n, basis_real, 'b-', label='实部 (cos)', linewidth=2)
    axes[i, 2].plot(n, basis_imag, 'r--', label='虚部 (sin)', linewidth=2)
    # 修改此处标题，去除LaTeX花括号，避免方框
    axes[i, 2].set_title(f'复指数基: k={k} (e^(i 2π {k} n / {N}))')
    axes[i, 2].set_xlabel('n')
    axes[i, 2].set_ylabel('幅度')
    axes[i, 2].legend()
    axes[i, 2].grid(True)

plt.tight_layout()
plt.show()

# 额外展示连续情况下的基函数
t = np.linspace(0, 1, 1000)
frequencies = [1, 2, 3, 5]

fig, axes = plt.subplots(len(frequencies), 2, figsize=(12, 10))

for i, f in enumerate(frequencies):
    # 正弦波
    axes[i, 0].plot(t, np.sin(2 * np.pi * f * t), 'b-', linewidth=2)
    axes[i, 0].set_title(f'连续正弦基: f = {f} Hz')
    axes[i, 0].set_xlabel('时间 (s)')
    axes[i, 0].set_ylabel('幅度')
    axes[i, 0].grid(True)
    
    # 余弦波
    axes[i, 1].plot(t, np.cos(2 * np.pi * f * t), 'r-', linewidth=2)
    axes[i, 1].set_title(f'连续余弦基: f = {f} Hz')
    axes[i, 1].set_xlabel('时间 (s)')
    axes[i, 1].set_ylabel('幅度')
    axes[i, 1].grid(True)

plt.tight_layout()
plt.show()
```

### 三维空间下的傅里叶基
在二维或三维空间下，傅里叶基函数不仅仅是一维的正弦/余弦波，而是空间中的波动模式。例如，三维傅里叶基可以写作：

$$
\psi_{\vec{k}}(\vec{r}) = e^{i \vec{k} \cdot \vec{r}}
$$

其中 $\vec{k}$ 是三维波矢，$\vec{r}$ 是空间坐标。其实部和虚部分别对应三维空间中的余弦和正弦波。下面用图像展示三维傅里叶基的实部（余弦波）和虚部（正弦波）。

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 三维网格
L = 2 * np.pi
n = 50
x = np.linspace(0, L, n)
y = np.linspace(0, L, n)
z = np.linspace(0, L, n)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')

# 选择一个三维波矢
kx, ky, kz = 1, 2, 3

# 三维傅里叶基函数
psi_real = np.cos(kx*X + ky*Y + kz*Z)
psi_imag = np.sin(kx*X + ky*Y + kz*Z)

# 取一个z截面进行可视化
z_index = n // 2
fig = plt.figure(figsize=(12,5))

ax1 = fig.add_subplot(1,2,1, projection='3d')
surf1 = ax1.plot_surface(X[:,:,z_index], Y[:,:,z_index], psi_real[:,:,z_index], cmap='viridis')
ax1.set_title('三维傅里叶基实部 (cos) 截面')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('幅度')
fig.colorbar(surf1, ax=ax1, shrink=0.5)

ax2 = fig.add_subplot(1,2,2, projection='3d')
surf2 = ax2.plot_surface(X[:,:,z_index], Y[:,:,z_index], psi_imag[:,:,z_index], cmap='plasma')
ax2.set_title('三维傅里叶基虚部 (sin) 截面')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('幅度')
fig.colorbar(surf2, ax=ax2, shrink=0.5)

plt.tight_layout()
plt.show()
```

### 一维信号的傅里叶基分解与详细例子

在一维信号处理中，任何有限长度的离散信号 $x[n]$ 都可以用一组正交的傅里叶基函数（正弦和余弦，或复指数）进行分解。这就是离散傅里叶变换（DFT）的本质。

**分解原理：**
- 对于长度为 $N$ 的信号 $x[n]$，其DFT为：
  $$
  X[k] = \sum_{n=0}^{N-1} x[n] e^{-i 2\pi k n / N}
  $$
- 反变换为：
  $$
  x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{i 2\pi k n / N}
  $$
- 其中 $e^{i 2\pi k n / N}$ 就是傅里叶基。

**例子：**
我们用一个简单的信号 $x[n] = \sin(2\pi f_1 n / N) + 0.5\cos(2\pi f_2 n / N)$，用DFT分解并重构它，并展示每个频率分量的幅度和相位。

```python
import numpy as np
import matplotlib.pyplot as plt

# 构造信号
N = 128
n = np.arange(N)
f1, f2 = 5, 20
x = np.sin(2 * np.pi * f1 * n / N) + 0.5 * np.cos(2 * np.pi * f2 * n / N)

# 傅里叶变换
X = np.fft.fft(x)
freqs = np.fft.fftfreq(N, d=1)

# 幅度和相位
amplitude = np.abs(X) / N
phase = np.angle(X)

# 重构信号
x_recon = np.fft.ifft(X).real

# 绘图
fig, axes = plt.subplots(4, 1, figsize=(10, 12))

axes[0].plot(n, x, label='原始信号')
axes[0].set_title('原始信号')
axes[0].set_xlabel('n')
axes[0].legend()

axes[1].stem(freqs[:N//2], amplitude[:N//2], basefmt=" ")
axes[1].set_title('频率分量幅度谱')
axes[1].set_xlabel('频率')
axes[1].set_ylabel('幅度')

axes[2].stem(freqs[:N//2], phase[:N//2], basefmt=" ")
axes[2].set_title('频率分量相位谱')
axes[2].set_xlabel('频率')
axes[2].set_ylabel('相位 (弧度)')

axes[3].plot(n, x_recon, label='重构信号', linestyle='--')
axes[3].plot(n, x, label='原始信号', alpha=0.5)
axes[3].set_title('傅里叶基重构信号')
axes[3].set_xlabel('n')
axes[3].legend()

plt.tight_layout()
plt.show()
```

#### 傅里叶基的数学叠加原理

对于长度为 $N$ 的离散信号 $x[n]$，傅里叶分解的数学本质是：

$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{i 2\pi k n / N}
$$

其中：
- $X[k]$ 是信号在第 $k$ 个频率上的傅里叶系数，表示该频率分量的“强度”和“相位”。
- $e^{i 2\pi k n / N}$ 是第 $k$ 个傅里叶基（复指数），也可以分解为正弦和余弦：
  $$
  e^{i 2\pi k n / N} = \cos\left(2\pi k n / N\right) + i \sin\left(2\pi k n / N\right)
  $$

**叠加过程：**
1. 每个 $k$，$X[k]$ 决定了该频率分量的幅度和相位。
2. 所有频率分量的基函数（正弦/余弦波）按照 $X[k]$ 的幅度和相位进行缩放和相移。
3. 把所有这些缩放和相移后的正弦/余弦波“叠加”起来，就还原出了原始信号。

**直观理解：**
- 就像拼积木一样，原始信号可以看作是不同频率、不同幅度、不同相位的正弦/余弦波的“和”。
- 低频分量决定信号的整体趋势，高频分量决定细节和快速变化。

**例子中的信号**

$$
x[n] = \sin\left(2\pi f_1 n / N\right) + 0.5 \cos\left(2\pi f_2 n / N\right)
$$

实际上就是 $f_1$ 和 $f_2$ 两个频率的傅里叶基的线性叠加。DFT 计算出的 $X[k]$ 只有在 $k=f_1$ 和 $k=f_2$ 附近有显著值，其余频率分量几乎为零。

**总结：**
- 傅里叶分解就是用一组正交的“基波”把信号分解、再叠加还原。
- 这组基波就是傅里叶基（正弦、余弦或复指数）。
- 每个基波的权重由 $X[k]$ 决定，所有基波加起来就是原始信号。

#### 傅里叶系数的可视化与解释

我们可以将傅里叶系数 $X[k]$ 以复平面上的点来展示，每个系数都包含幅度（模长）和相位（角度）。
- 幅度 $|X[k]|$ 表示该频率分量在信号中的强度。
- 相位 $\arg(X[k])$ 表示该分量的相位偏移。

下面用极坐标图（复平面）展示主要的傅里叶系数，并结合幅度谱和相位谱进行解释。

```python
import numpy as np
import matplotlib.pyplot as plt

# 复用前面例子的X, N, f1, f2
def plot_fourier_coeffs(X, N, f1, f2):
    k_values = [f1, f2]
    fig, ax = plt.subplots(1, 1, subplot_kw={'projection': 'polar'}, figsize=(7, 6))
    for k in k_values:
        r = np.abs(X[k]) / N
        theta = np.angle(X[k])
        ax.plot([theta], [r], 'o', label=f'k={k}, |X[k]|={r:.2f}, 相位={theta:.2f} rad')
        # 画出从原点到点的线
        ax.plot([0, theta], [0, r], '--', alpha=0.5)
    ax.set_title('主要傅里叶系数在复平面（极坐标）上的分布')
    ax.set_rmax(max(np.abs(X[k_values]) / N) * 1.2)
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
    plt.show()

# 调用可视化
plot_fourier_coeffs(X, N, f1, f2)

# 解释：
print('图中每个点代表一个主要频率分量的傅里叶系数，距离原点的长度表示幅度，角度表示相位。')
print('例如，k=5和k=20的点分别对应信号中的两个主频率分量。')
```

#### 为什么傅里叶基要带虚部？为什么虚部基也能组成实信号？

**1. 为什么傅里叶基要带虚部？**
- 傅里叶基 $e^{i2\pi k n/N}$ 是复指数函数，包含实部（余弦）和虚部（正弦）。
- 复指数的好处是：
  - 计算和推导更简洁，所有频率分量都能用一个公式统一表示。
  - 复数的乘法和旋转性质让频谱分析、卷积等操作变得非常自然。
- 复基函数的实部和虚部分别就是余弦和正弦：
  $$
  e^{i\theta} = \cos\theta + i\sin\theta
  $$

**2. 为什么虚部基也能组成实信号？**
- 虽然傅里叶分解用的是复数基，但对于实信号 $x[n]$，其傅里叶系数满足共轭对称：
  $$
  X[N-k] = X^*[k]
  $$
  这保证了所有虚部在叠加时会互相抵消，最终结果是实数。
- 具体来说：
  - 每一对共轭频率分量（$k$ 和 $N-k$）的虚部正好相反，实部相同。
  - 叠加时，虚部部分相加为零，实部部分相加为两倍余弦。
- 也可以用正弦和余弦基（实基）来分解信号，数学上等价。

**3. 总结**
- 复基只是数学上的简洁表达，最终实信号的虚部会自动抵消。
- 这也是为什么用复数傅里叶基分解实信号，最后重构出来的还是实信号。

**补充：**
- 如果信号本身是复数信号，傅里叶分解的结果就不再满足共轭对称，重构出来的信号也会有虚部。
